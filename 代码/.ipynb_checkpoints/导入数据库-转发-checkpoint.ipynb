```python
import pymysql
# å¼•å…¥pythonä¸­çš„tracebackæ¨¡å—ï¼Œè·Ÿè¸ªé”™è¯¯
import traceback
import sys


class MysqlUtil():
    def __init__(self):
        '''
            åˆå§‹åŒ–æ–¹æ³•ï¼Œè¿æ¥æ•°æ®åº“
        '''
        host = '192.168.8.126'  # ä¸»æœºå
        user = 'root'  # æ•°æ®åº“ç”¨æˆ·å
        password = '8008'  # æ•°æ®åº“å¯†ç 
        database = 'weibo'  # æ•°æ®åº“åç§°
        self.db = pymysql.connect(host=host, user=user, password=password, db=database)  # å»ºç«‹è¿æ¥
        self.cursor = self.db.cursor(cursor=pymysql.cursors.DictCursor)  # è®¾ç½®æ¸¸æ ‡ï¼Œå¹¶å°†æ¸¸æ ‡è®¾ç½®ä¸ºå­—å…¸ç±»å‹

    def insert(self, sql):
        '''
            æ’å…¥æ•°æ®åº“
            sql:æ’å…¥æ•°æ®åº“çš„sqlè¯­å¥
        '''
        try:
            # æ‰§è¡Œsqlè¯­å¥
            self.cursor.execute(sql)
            # æäº¤åˆ°æ•°æ®åº“æ‰§è¡Œ
            self.db.commit()
        except Exception:  # æ–¹æ³•ä¸€ï¼šæ•è·æ‰€æœ‰å¼‚å¸¸
            # å¦‚æœå‘ç”Ÿå¼‚å¸¸ï¼Œåˆ™å›æ»š
            print("å‘ç”Ÿå¼‚å¸¸", Exception)
            self.db.rollback()
        finally:
            # æœ€ç»ˆå…³é—­æ•°æ®åº“è¿æ¥
            self.db.close()

    def fetchone(self, sql):
        '''
            æŸ¥è¯¢æ•°æ®åº“ï¼šå•ä¸ªç»“æœé›†
            fetchone(): è¯¥æ–¹æ³•è·å–ä¸‹ä¸€ä¸ªæŸ¥è¯¢ç»“æœé›†ã€‚ç»“æœé›†æ˜¯ä¸€ä¸ªå¯¹è±¡
        '''
        try:
            # æ‰§è¡Œsqlè¯­å¥
            self.cursor.execute(sql)
            result = self.cursor.fetchone()
        except:  # æ–¹æ³•äºŒï¼šé‡‡ç”¨tracebackæ¨¡å—æŸ¥çœ‹å¼‚å¸¸
            # è¾“å‡ºå¼‚å¸¸ä¿¡æ¯
            traceback.print_exc()
            # å¦‚æœå‘ç”Ÿå¼‚å¸¸ï¼Œåˆ™å›æ»š
            self.db.rollback()
            result = 'None'

        finally:
            # æœ€ç»ˆå…³é—­æ•°æ®åº“è¿æ¥
            self.db.close()
        return result

    def fetchall(self, sql):
        '''
            æŸ¥è¯¢æ•°æ®åº“ï¼šå¤šä¸ªç»“æœé›†
            fetchall(): æ¥æ”¶å…¨éƒ¨çš„è¿”å›ç»“æœè¡Œ.
        '''
        try:
            # æ‰§è¡Œsqlè¯­å¥
            self.cursor.execute(sql)
            results = self.cursor.fetchall()
        except:  # æ–¹æ³•ä¸‰ï¼šé‡‡ç”¨sysæ¨¡å—å›æº¯æœ€åçš„å¼‚å¸¸
            # è¾“å‡ºå¼‚å¸¸ä¿¡æ¯
            info = sys.exc_info()
            print(info[0], ":", info[1])
            # å¦‚æœå‘ç”Ÿå¼‚å¸¸ï¼Œåˆ™å›æ»š
            self.db.rollback()
            results = 'None'
        finally:
            # æœ€ç»ˆå…³é—­æ•°æ®åº“è¿æ¥
            self.db.close()
        return results

    def delete(self, sql):
        '''
            åˆ é™¤ç»“æœé›†
        '''
        try:
            # æ‰§è¡Œsqlè¯­å¥
            self.cursor.execute(sql)
            self.db.commit()
        except:  # æŠŠè¿™äº›å¼‚å¸¸ä¿å­˜åˆ°ä¸€ä¸ªæ—¥å¿—æ–‡ä»¶ä¸­ï¼Œæ¥åˆ†æè¿™äº›å¼‚å¸¸
            # å°†é”™è¯¯æ—¥å¿—è¾“å…¥åˆ°ç›®å½•æ–‡ä»¶ä¸­
            f = open("\log.txt", 'a')
            traceback.print_exc(file=f)
            f.flush()
            f.close()
            # å¦‚æœå‘ç”Ÿå¼‚å¸¸ï¼Œåˆ™å›æ»š
            self.db.rollback()
        finally:
            # æœ€ç»ˆå…³é—­æ•°æ®åº“è¿æ¥
            self.db.close()

    def update(self, sql):
        '''
            æ›´æ–°ç»“æœé›†
        '''
        try:
            # æ‰§è¡Œsqlè¯­å¥
            self.cursor.execute(sql)
            self.db.commit()
        except:
            # å¦‚æœå‘ç”Ÿå¼‚å¸¸ï¼Œåˆ™å›æ»š
            self.db.rollback()
        finally:
            # æœ€ç»ˆå…³é—­æ•°æ®åº“è¿æ¥
            self.db.close()
```

```python
import requests
from urllib.parse import urlencode
import time
import random
from pyquery import PyQuery as pq
import sys

# è®¾ç½®ä»£ç†ç­‰ï¼ˆæ–°æµªå¾®åšçš„æ•°æ®æ˜¯ç”¨ajaxå¼‚æ­¥ä¸‹æ‹‰åŠ è½½çš„ï¼Œnetwork->xhrï¼‰
host = 'm.weibo.cn'
base_url = 'https://%s/api/container/getIndex?' % host
user_agent = 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Mobile Safari/537.36'

# è®¾ç½®è¯·æ±‚å¤´
headers = {
    'Host': host,
    'keep': 'close',
    # 'Referer': 'https://m.weibo.cn/search?containerid=231522type%3D1%26t%3D10%26q%3D%23%E5%A6%82%E4%BD%95%E7%9C%8B%E5%BE%85%E5%8F%8D%E5%86%85%E5%8D%B7%E7%83%AD%E6%BD%AE%23&extparam=%23%E5%A6%82%E4%BD%95%E7%9C%8B%E5%BE%85%E5%8F%8D%E5%86%85%E5%8D%B7%E7%83%AD%E6%BD%AE%23&luicode=10000011&lfid=100103type%3D38%26q%3D%E5%86%85%E5%8D%B7%26t%3D0',
    'User-Agent': user_agent
}

save_per_n_page = 10

from datetime import datetime


def time_formater(input_time_str):
    input_format = '%a %b %d %H:%M:%S %z %Y'
    output_format = '%Y-%m-%d %H:%M:%S'

    return datetime.strptime(input_time_str, input_format).strftime(output_format)


# æŒ‰é¡µæ•°æŠ“å–æ•°æ®
def get_single_page(page, keyword):
    # è¯·æ±‚å‚æ•°
    params = {
        'containerid': f'100103type=1&q=#{keyword}#',  # ã€ã€æ•™è‚²å†…å·ã€èŒåœºå†…å·ã€å¦‚ä½•çœ‹å¾…å†…å·çš„ç¤¾ä¼šçŠ¶æ€ã€å¦‚ä½•é¿å…å©šå§»å†…å·ã€
        'page_type': 'searchall',
        'page': page
    }
    url = base_url + urlencode(params)
    print(url)
    error_times = 3
    while True:
        response = requests.get(url, headers=headers)  # ,proxies=abstract_ip.get_proxy()
        if response.status_code == 200:
            if len(response.json().get('data').get('cards')) > 0:
                return response.json()
        time.sleep(3)
        error_times += 1
        # è¿ç»­å‡ºé”™æ¬¡æ•°è¶…è¿‡ 3
        if error_times > 3:
            return None


# é•¿æ–‡æœ¬çˆ¬å–ä»£ç æ®µ
def getLongText(lid):  # lidä¸ºé•¿æ–‡æœ¬å¯¹åº”çš„id
    # é•¿æ–‡æœ¬è¯·æ±‚å¤´
    headers_longtext = {
        'Host': host,
        'Referer': 'https://m.weibo.cn/status/' + lid,
        'User-Agent': user_agent
    }
    params = {
        'id': lid
    }
    url = 'https://m.weibo.cn/statuses/extend?' + urlencode(params)
    try:
        response = requests.get(url, headers=headers_longtext)  # proxies=abstract_ip.get_proxy()
        if response.status_code == 200:  # æ•°æ®è¿”å›æˆåŠŸ
            jsondata = response.json()
            tmp = jsondata.get('data')
            return pq(tmp.get("longTextContent")).text()  # è§£æè¿”å›ç»“æ„ï¼Œè·å–é•¿æ–‡æœ¬å¯¹åº”å†…å®¹
    except:
        pass


# è§£æé¡µé¢è¿”å›çš„jsonæ•°æ®
count = 0

'''
ä¿®æ”¹åçš„é¡µé¢çˆ¬å–è§£æå‡½æ•°
'''


def parse_page(json_data):
    global count
    items = json_data.get('data').get('cards')

    for index, item in enumerate(items):
        if item.get('card_type') == 7:
            print('å¯¼è¯­')
            continue
        elif item.get('card_type') == 8 or (item.get('card_type') == 11 and item.get('card_group') is None):
            continue
        # topic = json_data.get('data').get('cardlistInfo').get('cardlist_head_cards')[0]
        # # å•ç‹¬çš„å…³é”®è¯æŠ“å–ä¸æ˜¯è¶…è¯ï¼Œä¼šæœ‰ topic == null
        # if topic is None or topic.get('head_data', None) is None:
        #     topic = keyword
        # else:
        #     topic = topic.get('head_data').get('title')
        if item.get('mblog', None):
            item = item.get('mblog')
        else:
            item = item.get('card_group')[0].get('mblog')
        if item:
            if item.get('isLongText') is False:  # ä¸æ˜¯é•¿æ–‡æœ¬
                data = {
                    'wid': item.get('id'),
                    'user_name': item.get('user').get('screen_name'),
                    'user_id': item.get('user').get('id'),
                    'gender': item.get('user').get('gender'),
                    'publish_time': time_formater(item.get('created_at')),
                    'text': pq(item.get("text")).text(),  # ä»…æå–å†…å®¹ä¸­çš„æ–‡æœ¬
                    'like_count': item.get('attitudes_count'),  # ç‚¹èµæ•°
                    'comment_count': item.get('comments_count'),  # è¯„è®ºæ•°
                    'forward_count': item.get('reposts_count'),  # è½¬å‘æ•°
                }
            else:  # é•¿æ–‡æœ¬æ¶‰åŠæ–‡æœ¬çš„å±•å¼€
                tmp = getLongText(item.get('id'))  # è°ƒç”¨å‡½æ•°
                data = {
                    'wid': item.get('id'),
                    'user_name': item.get('user').get('screen_name'),
                    'user_id': item.get('user').get('id'),
                    'gender': item.get('user').get('gender'),
                    'publish_time': time_formater(item.get('created_at')),
                    'text': tmp,  # ä»…æå–å†…å®¹ä¸­çš„æ–‡æœ¬
                    'like_count': item.get('attitudes_count'),
                    'comment_count': item.get('comments_count'),
                    'forward_count': item.get('reposts_count'),
                }
            count += 1
            print(f'total count: {count}')
            yield data


import os, csv

if __name__ == '__main__':
    keyword = 'å¼ æ–°æˆ å—ä¼¤'
    keyword = sys.argv[1]
    print (keyword)
    sys.exit()
    result_file = f'{keyword}.csv'
    if not os.path.exists(result_file):
        with open(result_file, mode='w', encoding='utf-8-sig', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(['wid', 'user_name', 'user_id', 'gender',
                             'publish_time', 'text', 'like_count', 'comment_count',
                             'forward_count'])

    temp_data = []

    empty_times = 0

    for page in range(1, 50000):  # ç€‘å¸ƒæµä¸‹æ‹‰å¼ï¼ŒåŠ è½½
        print(f'page: {page}')
        json_data = get_single_page(page, keyword)
        if json_data == None:
            print('json is none')
            break

        if len(json_data.get('data').get('cards')) <= 0:
            empty_times += 1
        else:
            empty_times = 0
        if empty_times > 3:
            print('\n\n consist empty over 3 times \n\n')
            break

        for result in parse_page(json_data):  # éœ€è¦å­˜å…¥çš„å­—æ®µ
            temp_data.append(result)
        if page % save_per_n_page == 0:
            with open(result_file, mode='a+', encoding='utf-8-sig', newline='') as f:
                writer = csv.writer(f)
                for d in temp_data:
                    writer.writerow(
                        [d['wid'], d['user_name'], d['user_id'], d['gender'],
                         d['publish_time'], d['text'], d['like_count'], d['comment_count'],
                         d['forward_count']])
            print(f'\n\n------cur turn write {len(temp_data)} rows to csv------\n\n')
            sys.exit()
            temp_data = []
        time.sleep(random.randint(2, 6))  # çˆ¬å–æ—¶é—´é—´éš”
```
